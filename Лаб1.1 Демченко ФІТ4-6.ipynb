```python
# 1. Побудувати горизонтальну стовпчикову діаграму, на якій відображаються 10 
# найпопулярніших пісень на Spotify у 2023 році за кількістю 
# прослуховувань.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("spotify-2023.csv", encoding='latin1')

# Перетворення колонки 'streams' у числовий тип даних
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')

top_10_songs = df.nlargest(10, 'streams')
plt.figure(figsize=(10, 6))
plt.barh(top_10_songs['track_name'], top_10_songs['streams'], color='skyblue')
plt.xlabel('Прослуховування')
plt.ylabel('Назва пісні')
plt.title('10 найпопулярніших пісень на Spotify у 2023 році')
plt.gca().invert_yaxis()
plt.show()
```

```python
# Побудувати кругову діаграму для тих самих даних.
```

```python
plt.figure(figsize=(8, 8))
plt.pie(top_10_songs['streams'], labels=top_10_songs['track_name'], autopct='%1.1f%%', startangle=140)
plt.title('10 найпопулярніших пісень на Spotify у 2023 році')
plt.axis('equal')
plt.show()
```

```python
# 2. Створіть кругову діаграму або секторну діаграму, яка відображає розподіл
# за danceability серед найпопулярніших пісень у 2023 році.
```

```python
plt.figure(figsize=(8, 8))
plt.pie(top_10_songs['danceability_%'], labels=top_10_songs['track_name'], autopct='%1.1f%%', startangle=140)
plt.title('Розподіл за danceability серед найпопулярніших пісень у 2023 році')
plt.axis('equal')
plt.show()
```

```python
# 3. Побудуйте лінійний графік, який відображає зміни кількості прослуховувань 
# певного треку протягом року 2023 року.
```

```python
song_1 = df[df['track_name'] == 'Sunflower - Spider-Man: Into the Spider-Verse']
plt.figure(figsize=(10, 6))
plt.plot(song_1['released_month'], song_1['streams'], marker='o')
plt.xlabel('Місяць релізу')
plt.ylabel('Прослуховування')
plt.title('Зміни кількості прослуховувань пісні "Sunflower - Spider-Man: Into the Spider-Verse" у 2023 році')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
#  Створіть стовпчикову діаграму, на якій відображаються 10 найпопулярніших 
# артистів на Spotify у 2023 році за кількістю слухачів
```

```python
top_10_artists = df.groupby('artist(s)_name')['streams'].sum().nlargest(10)
plt.figure(figsize=(10, 6))
plt.bar(top_10_artists.index, top_10_artists.values, color='skyblue')
plt.xlabel('Виконавець')
plt.ylabel('Прослуховування')
plt.title('10 найпопулярніших артистів на Spotify у 2023 році за кількістю прослуховувань')
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. Побудуйте горизонтальну стовпчикову діаграму, яка порівнює кількість 
# прослуховувань трьох конкретних треків у 2023 році
```

```python
songs_to_compare = df[df['track_name'].isin(['Daylight', 'Viva La Vida', 'Sweater Weather'])]
plt.figure(figsize=(10, 6))
plt.barh(songs_to_compare['track_name'], songs_to_compare['streams'], color=['skyblue', 'green', 'red'])
plt.xlabel('Прослуховування')
plt.ylabel('Назва пісні')
plt.title('Порівняння кількості прослуховувань трьох пісень у 2023 році')
plt.gca().invert_yaxis()
plt.show()
```
