```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("shopping_trends.csv")


print("Розмір таблиці:")
print(df.shape)

print("\nПропуски в даних:")
print(df.isnull().sum())

print("\nКількість дублікатів:")
print(df.duplicated().sum())

print("\nТипи даних стовпців:")
print(df.dtypes)

df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
# 1 Гістограма віку (age):
# Побудуйте гістограму віку клієнтів, використовуючи Seaborn
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True, color='skyblue')
plt.title('Гістограма віку клієнтів')
plt.xlabel('Вік')
plt.ylabel('Частота')
plt.show()
```

```python
#  2. Порівняння розподілу покупок за статтею (item_purchased):
# Створіть стовпчасту діаграму або графік розподілу, щоб порівняти, які
# товари покупають клієнти найчастіше.

```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='item_purchased', data=df, order=df['item_purchased'].value_counts().index, palette='viridis')
plt.title('Розподіл покупок за статтею')
plt.xlabel('Кількість покупок')
plt.ylabel('Стаття товару')
plt.show()
```

```python
# 3. Залежність між віком та сумою покупок (purchase_amount):
# Використовуйте scatter plot, щоб візуалізувати залежність між віком та
# сумою покупок.
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='purchase_amount_(usd)', data=df, color='orange')
plt.title('Залежність між віком та сумою покупок')
plt.xlabel('Вік')
plt.ylabel('Сума покупки')
plt.show()
```

```python
# 4. Графік розподілу оцінок товарів (review_rating):
# Побудуйте графік розподілу оцінок товарів для оцінки якості.
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['review_rating'], bins=5, kde=True, color='green')
plt.title('Розподіл оцінок товарів')
plt.xlabel('Оцінка')
plt.ylabel('Частота')
plt.show()
```

```python
# 5. Категоріальна діаграма для статі (gender):
# Створіть категоріальну діаграму, щоб відобразити розподіл покупок за
# статтею.
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='gender', data=df, palette='Set2')
plt.title('Розподіл покупок за статтю')
plt.xlabel('Стать')
plt.ylabel('Кількість покупок')
plt.show()
```

```python
# 6. Залежність між рейтингом і сумою покупок (review_rating і
# purchase_amount):
# Вивчіть, чи існує зв'язок між рейтингом товару і сумою покупок за
# допомогою scatter plot
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='review_rating', y='purchase_amount_(usd)', data=df, hue='gender')
plt.title('Залежність між рейтингом і сумою покупок')
plt.xlabel('Оцінка товару')
plt.ylabel('Сума покупки')
plt.show()
```

```python
# 7. Сезонність покупок (season):
# Побудуйте стовпчату діаграму, яка відображає, в який сезон клієнти
# роблять найбільше покупок.
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='season', data=df, palette='coolwarm')
plt.title('Сезонність покупок')
plt.xlabel('Сезон')
plt.ylabel('Кількість покупок')
plt.show()
```

```python
# 8. Графік розподілу розмірів одягу (size):
# Створіть графік розподілу розмірів одягу, щоб визначити популярність
# різних розмірів.
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='size', data=df, order=df['size'].value_counts().index, palette='magma')
plt.title('Розподіл розмірів одягу')
plt.xlabel('Розмір')
plt.ylabel('Кількість покупок')
plt.show()
```

```python
# 9. Кількість покупок відносно типу оплати (payment_method):
# Порівняйте, скільки покупок робиться за різними методами оплати за
# допомогою стовпчатої діаграми.
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='payment_method', data=df, order=df['payment_method'].value_counts().index, palette='Set3')
plt.title('Кількість покупок за типом оплати')
plt.xlabel('Кількість покупок')
plt.ylabel('Тип оплати')
plt.show()
```

```python
# 10.Залежність між кількістю попередніх покупок (previous_purchases) і
# сумою поточної покупки (purchase_amount):
# Створіть scatter plot для визначення, чи існує зв'язок між кількістю
# попередніх покупок та сумою поточної покупки
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='previous_purchases', y='purchase_amount_(usd)', data=df, hue='payment_method')
plt.title('Залежність між кількістю попередніх покупок і сумою покупки')
plt.xlabel('Кількість попередніх покупок')
plt.ylabel('Сума покупки')
plt.show()
```

```python
# Завдання 2
# Побудуйте графік поверхні (3D surface plot) для візуалізації залежності
# між трьома числовими змінними з використанням бібліотеки Seaborn.
# Налаштуйте відображення графіка, включаючи назву графіку, підписи
# осей і інші необхідні параметри для кращого розуміння візуалізації.
# Додайте кольорову легенду 𝑧 = 𝑠𝑖𝑛 √𝑥^2 + 𝑦^2, 𝑥 ∈ [−5; 5], 𝑦 ∈ [−5; 5]
```

```python
import numpy as np

from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, y)

Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

surface = ax.plot_surface(X, Y, Z, cmap='viridis')

fig.colorbar(surface)

ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
